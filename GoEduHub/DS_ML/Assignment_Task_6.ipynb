#### 1. Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a carâ€™s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.
         Days=[1,2,3,4,5,6,7,8]
         Speed=[60,62,61,58,56,57,46,63]

```python
import numpy as np
import matplotlib.pyplot as plt
 
Days=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]
plt.figure(figsize=(10,5))
plt.plot(Days,Speed,color='blue',linewidth=2)
plt.title('Car speed')
plt.xlabel('Days')
plt.ylabel('Speed')
plt.show()
```

#### 2. Now to above car data apply some string formats  like line style example green dotted line, marker shape like +, change markersize, markerface color etc


```python
plt.figure(figsize=(10,5))
plt.plot(Days,Speed,color='green',linewidth=3,linestyle='--',marker='+',markersize='16',markeredgewidth=2,markeredgecolor='red')
plt.title('Car speed')
plt.xlabel('Days')
plt.ylabel('Speed')
plt.show()
```

#### 3. Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed in 8 days.

```python
days=[1,2,3,4,5,6,7,8]
max_speed=[80,91,92,88,77,79,76,75]
min_speed=[42,43,40,42,33,36,34,35]
avg_speed=[46,58,57,56,40,42,41,36]
plt.plot(days,max_speed,color='red',linewidth=2,label='max_speed')
plt.plot(days,min_speed,color='green',linewidth=4,label='min_speed')
plt.plot(days,avg_speed,color='blue',linewidth=3,linestyle='--',label='avg_speed')
plt.title('Car speed')
plt.xlabel('Days')
plt.ylabel('Speed')
plt.legend()
plt.grid()
plt.show()
```

#### 4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,color='red',label='sin x')
plt.plot(x,z,color='blue',label='cos x')

plt.legend()
plt.grid()
plt.show()
```

#### 5. Plot Simple bar chart showing popularity of Programming Languages.

Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]

Security = [44 ,36 ,55, 50, 42]

Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.

```python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]
x_axis=np.arange(len(Languages))
plt.bar(x_axis-0.2,Popularity,0.4,color='red',label='Popularity')
plt.bar(x_axis+0.2,Security,0.4,color='orange',label='Security')
plt.title('Popularity and Security - Programming Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity & Security')
plt.legend()
plt.grid()
plt.show()
```

```python
x_axis=np.arange(len(Languages))
plt.barh(x_axis-0.2,Popularity,0.4,color='orange',label='Popularity')
plt.barh(x_axis+0.2,Security,0.4,color='purple',label='Security')
plt.title('Popularity and Security of Programming Languages(Horizontal)')
plt.ylabel('Programming Languages')
plt.xlabel('Popularity and Security')
plt.legend()
plt.grid()
plt.show()
```

#### 7.Titanic Data Set Download Data

 Load the data file

 (i) Create a pie chart presenting the male/female proportion 

 (ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd

data=pd.read_csv('Titanic-Dataset.csv')
data_1=data['Sex'].value_counts()
print(data_1)
plt.pie(data_1,colors=['orange','blue'],labels=['male','female'],autopct='%1.1f%%')
plt.title('male/female proportion',fontsize=16)
plt.legend()
plt.show()
```

```python
plt.scatter(data['Fare'],data['Age'],c=pd.factorize(data['Sex'])[0],alpha=1)
plt.title('Fare paid & age scatter plot',fontsize=16)
plt.xlabel('Fare paid')
plt.ylabel('Age')
plt.show()
```
